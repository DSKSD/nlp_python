#### Fine-Grained Selection of Words

이번엔 좀 긴 단어들을 살펴보자, 어쩌면 이들은 좀더 특징적이고 informative할지도 모른다. 이를 위해 우리는 집합의 개념을 적용할 것이다. <우리는 15자가 넘는 단어들을 뽑고 싶다.>

```python
V = set(text1)
```

```python
long_words = [w for w in V if len(w) > 15]
```

```python
sorted(long_words)
```

### 1.5 Automatic Natural Language Understanding

인공지능 분야에서 중요한 파트 중 하나는 'understanding language'이다. 이를 위해서는 아주 폭넓은 language processing tasks(information extraction, inference, and summarization 등)이 필요하다.

몇 년간 이러한 목표는 아주 어려움을 겪었으나 NLP 기술들이 성숙하면서 여러 Robust methods들이 등장했다. 지금부터는 당신의 흥미를 불러일으키기 위해 이러한 방법들을 소개하겠다.

#### Word Sense Disambiquation 

주어진 맥락에서 어떠한 단어가 의미하는 바가 무엇인지 알고자 하는 방법

ex>

 * a. serve: help with food or drink; hold an office; put ball into play 
 * b. dish: plate; course of a meal; communications device


'he served the dish' 

-> 두 단어 모두 food와 관련하여 사용함을 알 수 있음

또 다른 예로 많은 뜻을 가진 'by'

* a. The lost children were found by the searchers (agentive) 
* b. The lost children were found by the mountain (locative) 
* c. The lost children were found by the afternoon (temporal) 

#### Pronoun Resolution 

동사의 주어와 목적어를 탐지하는 방법.

당신은 이것을 초등학교 때 배우지만 생각보다 어렵다. 

ex> 'the thieves stole the paintings' 이 문장을 보면 누가 훔쳤는지 쉽게 말할 수 있다. 

하지만 뒤에 따라오는 문장들을 고려해보자. 

 * a. The thieves stole the paintings. They were subsequently sold. 
 * b. The thieves stole the paintings. They were subsequently caught. 
 * c. The thieves stole the paintings. They were subsequently found.

뒷따라 오는동작의 주체가 무엇인지 알아내보자. 

이 질문의 답을 하기 위해서는 먼저 'they'의 선행자(antecedent)를 찾아야 한다. (thieves or paintings)

이러한 문제를 해결하는 Computational techniques는 <strong>anaphora resolution</strong>(identifying what a pronoun or noun phrase refers to)와 <strong>semantic role labeling</strong>(identifying how a noun phrase relates to the verb(as agent, patient, instrument, and so on.)) 등을 포함한다.

#### Generating Language Output

만약 우리가 language understanding의 문제를 해결했다면 그 다음으로 language output을 만들어내는 단계로 갈 수 있다. <strong>question answering</strong> and <strong>machine translation</strong>

 * Text: ... The thieves stole the paintings. They were subsequently sold. ... 
 * Human: Who or what was sold? 
 * Machine: The paintings.
